### MISSING VALUE TREATMENT

```python
# for the categorical/ discrete  use to fill with mode 
# while numerical we use to fill with mean or median
# let use KNNIputer
```

```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include="number").columns:
    df[i]=impute.fit_transform(df[[i]])
```
